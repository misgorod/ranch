```python
import csv

def to_float(s):
    try:
        s=float(s)
    except ValueError:
        pass    
    return s

metrics = []

with open('metrics/ranch-metrics.csv', 'r') as f:
    r = csv.DictReader(f)
    for row in r:
        metrics.append(dict(row))
    
for metric in metrics:
    for key in metric:
        metric[key] = to_float(metric[key])
```
